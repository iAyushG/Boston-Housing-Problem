```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error
```

```python
#generate 'n' evenly spaces values from zero radians to 2 PI radians
n = 200
x = np.linspace(0, 2 * np.pi, n)
sine_values = np.sin(x)

#plotting the sin wave
plt.plot(x, sine_values)
```

```python
#adding random noise to the sine values
noise = 0.5
noisy_sine_values = sine_values + np.random.uniform(-noise, noise, n)

#plotting noisy data and the normal data together
plt.plot(x, noisy_sine_values, color='r')
plt.plot(x, sine_values, linewidth=3)
```

```python
#MSE using equation
error_value = (1/n) * sum(np.power(sine_values - noisy_sine_values, 2))
error_value
```

```python
#MSE using mean_squared_error function from sklearn library
mean_squared_error(sine_values, noisy_sine_values)
```
